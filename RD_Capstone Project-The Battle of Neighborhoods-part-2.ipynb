```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# assign relevant part of JSON to venues
venues = results['response']['groups'][0]['items']
leeds_nearby_venues = json_normalize(venues) # flatten JSON
```

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
leeds_nearby_venues =leeds_nearby_venues.loc[:, filtered_columns]
# filter the category for each row
leeds_nearby_venues['venue.categories'] = leeds_nearby_venues.apply(get_category_type, axis=1)
# clean columns
leeds_nearby_venues.columns = [col.split(".")[-1] for col in leeds_nearby_venues.columns]

leeds_nearby_venues.shape
```
